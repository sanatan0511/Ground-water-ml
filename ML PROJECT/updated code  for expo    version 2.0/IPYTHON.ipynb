```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from statsmodels.tsa.arima.model import ARIMA
import warnings
import random
from deap import base, creator, tools, algorithms


import webview


webview.create_window("Bhuvan GWIS", "https://bhuvan-app1.nrsc.gov.in/gwis/gwis.php")


webview.start()

print("Thank! for visit please visit bhuvan isro  website for  detail satellite view: https://bhuvan-app1.nrsc.gov.in/gwis/gwis.php" )

import pyttsx3

engine = pyttsx3.init()

voices = engine.getProperty('voices')


female_voice = None
for voice in voices:
    if 'female' in voice.name.lower():
        female_voice = voice
        break


if female_voice:
    engine.setProperty('voice', female_voice.id)
else:
    print("No female voice found. Using default voice.")

rate = engine.getProperty('rate')
engine.setProperty('rate', 125)

engine.setProperty('volume', 1.0)

text = "Welcome to BHUMI AI BHUMI AI welcomes You!, Dear User, For security purposes, never share any confidential information Be alert, be safe  and Always planting  more plants remember one step save toward change enviroment save ecosystem and keep ecosystem clean,friendly Thank you!"

engine.say(text)
engine.runAndWait()

years = ["2023-24", "2022-23", "2021-22", "2019-20", "2016-17"]
data_files = ["2023data.csv", "2022-23data.csv", "2021-22data.csv", "2019-20data.csv", "2016-17data.csv"]
df_list = []

dataframes = {}

for year, file in zip(years, data_files):
    dataframes[year] = pd.read_csv(file)
    print(f"Loaded {year} data: {dataframes[year].shape}")

# Combine data for trend analysis
groundwater_data = pd.concat(dataframes.values(), keys=years, names=['Year'])

# Reset index for easier manipulation
groundwater_data.reset_index(inplace=True)

# Graphical Analysis
def visualize_trends(df):
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=df, x='Year', y='Ground Water Extraction for all uses (ha.m)', hue='STATE', estimator='mean')
    plt.xticks(rotation=45)
    plt.title('Groundwater Extraction Trends Across States')
    plt.ylabel('Extraction (ha.m)')
    plt.xlabel('Year')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()

visualize_trends(groundwater_data)

for file in data_files:
    df_temp = pd.read_csv(file)
    df_temp['Year'] = file.split('data')[0]
    df_list.append(df_temp)
df = pd.concat(df_list, ignore_index=True)


state_input = input("Enter state (or leave blank for overall data): ").strip().upper()
district_input = input("Enter district (or leave blank for overall data): ").strip().upper()

if state_input:
    df['STATE'] = df['STATE'].str.upper()
if district_input:
    df['DISTRICT'] = df['DISTRICT'].str.upper()

if state_input and district_input:
    filtered_df = df[(df['STATE'] == state_input) & (df['DISTRICT'] == district_input)]
elif state_input:
    filtered_df = df[df['STATE'] == state_input]
elif district_input:
    filtered_df = df[df['DISTRICT'] == district_input]
else:
    filtered_df = df.copy()

if filtered_df.empty:
    print("No data found for the given state/district. Exiting.")
    exit()
else:
    print("Filtered data sample:")
    print(filtered_df.head())


exclude_cols = ['S.No', 'STATE', 'DISTRICT', 'Year']
numeric_cols = [col for col in filtered_df.select_dtypes(include=[np.number]).columns if col not in exclude_cols]

for col in numeric_cols:
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    sns.histplot(filtered_df[col].dropna(), kde=True)
    plt.title(f'Histogram of {col}')
    
    plt.subplot(1, 2, 2)
    sns.boxplot(x=filtered_df[col])
    plt.title(f'Boxplot of {col}')
    
    plt.tight_layout()
    plt.show()
# fea_tures = ['Stage of Ground Water Extraction (%)', 
#             'Net Annual Ground Water Availability for Future Use (ham))', 
#             'Rainfall(mm)','Ground Water Recharge (ham)','Annual Extractable Ground water Resource (ham)']






features = ['Stage of Ground Water Extraction (%)', 
            'Net Annual Ground Water Availability for Future Use (ham))', 
            'Rainfall(mm)']

for feature in features:
    mean_value = filtered_df[feature].replace(0, np.nan).mean()
    mean_value = 0.1 if np.isnan(mean_value) or mean_value == 0 else mean_value
    filtered_df[feature] = filtered_df[feature].replace(0, mean_value).fillna(mean_value)

risk_df = filtered_df.dropna(subset=features).copy()
if risk_df.empty:
    print("No data left after filtering. Exiting.")
    exit()


def assign_risk(row):
    stage = row['Stage of Ground Water Extraction (%)']
    availability = row['Net Annual Ground Water Availability for Future Use (ham))']
    if stage > 70 or availability < 20000:
        return 'High'
    elif stage > 50 or availability < 40000:
        return 'Medium'
    else:
        return 'Low'

risk_df['RiskLabel'] = risk_df.apply(assign_risk, axis=1)


X = risk_df[features].values
y = risk_df['RiskLabel'].values

scaler_features = MinMaxScaler()
X_scaled = scaler_features.fit_transform(X)

le = LabelEncoder()
y_encoded = le.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

y_pred = rf_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy * 100:.2f}%")
print("Classification Report:\n", report)

risk_pred_class = rf_model.predict(X_scaled)
risk_pred_label = le.inverse_transform(risk_pred_class)
risk_df['PredictedRisk'] = risk_pred_label

print("\nAdvanced Risk Classification Results:")
print(risk_df[['STATE', 'DISTRICT', 'Year'] + features + ['RiskLabel', 'PredictedRisk']].head())

extraction = ctrl.Antecedent(np.arange(0, 101, 1), 'extraction')
availability = ctrl.Antecedent(np.arange(0, 100001, 1000), 'availability')
risk_output = ctrl.Consequent(np.arange(0, 11, 1), 'risk_output')


extraction['low'] = fuzz.trimf(extraction.universe, [0, 0, 50])
extraction['medium'] = fuzz.trimf(extraction.universe, [30, 50, 70])
extraction['high'] = fuzz.trimf(extraction.universe, [60, 100, 100])

availability['low'] = fuzz.trimf(availability.universe, [0, 0, 40000])
availability['medium'] = fuzz.trimf(availability.universe, [30000, 50000, 70000])
availability['high'] = fuzz.trimf(availability.universe, [60000, 100000, 100000])

risk_output['low'] = fuzz.trimf(risk_output.universe, [0, 0, 5])
risk_output['medium'] = fuzz.trimf(risk_output.universe, [3, 5, 7])
risk_output['high'] = fuzz.trimf(risk_output.universe, [5, 10, 10])


rule1 = ctrl.Rule(extraction['high'] | availability['low'], risk_output['high'])
rule2 = ctrl.Rule(extraction['medium'] & availability['medium'], risk_output['medium'])
rule3 = ctrl.Rule(extraction['low'] & availability['high'], risk_output['low'])


risk_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])
risk_simulation = ctrl.ControlSystemSimulation(risk_ctrl)


mean_extraction = filtered_df['Stage of Ground Water Extraction (%)'].mean()
mean_availability = filtered_df['Net Annual Ground Water Availability for Future Use (ham))'].mean()

risk_simulation.input['extraction'] = mean_extraction
risk_simulation.input['availability'] = mean_availability
risk_simulation.compute()

print("\nFuzzy Logic Risk Output Score:", risk_simulation.output['risk_output'])


extraction.view()
plt.show()

availability.view()
plt.show()

risk_output.view(sim=risk_simulation)
plt.show()



risk_counts = risk_df['PredictedRisk'].value_counts()
if 'High' in risk_counts and risk_counts['High'] > 0:
    overall_risk = 'High'
elif 'Medium' in risk_counts and risk_counts['Medium'] > 0:
    overall_risk = 'Medium'
else:
    overall_risk = 'Low'

if overall_risk == 'High':
    final_suggestion = ("Urgent action required. Implement strict water regulations, "
                        "increase recharge, and invest in alternative water sources. planting trees like Neem,Ashoka,tulsi,banana etc")
elif overall_risk == 'Medium':
    final_suggestion = ("Monitor groundwater levels and adopt conservation measures. "
                        "Plan for future water demand carefully.plant tree like  Banyan tree")
else:
    final_suggestion = ("Groundwater levels are stable; maintain current practices and monitor periodically.plant trees like Tulsi,Neem etc")

print("\nFinal Risk Assessment and Suggestions:")
print("Overall Risk Level:", overall_risk)
print("Final Suggestion:", final_suggestion)





def forecast_arima(series, forecast_periods=15, order=(1,1,1)):
    model = ARIMA(series, order=order)
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=forecast_periods)
    return forecast

def groundwater_forecast(filtered_df, forecast_periods=15):
    
    filtered_df['Year'] = pd.to_numeric(filtered_df['Year'], errors='coerce')

   
    years = sorted(filtered_df['Year'].dropna().unique())
    future_years = list(range(int(years[-1]) + 1, int(years[-1]) + forecast_periods + 1))

    
    extraction_ts = filtered_df.groupby('Year')['Stage of Ground Water Extraction (%)'].mean()
    availability_ts = filtered_df.groupby('Year')['Net Annual Ground Water Availability for Future Use (ham)'].mean()

  
    extraction_forecast = forecast_arima(extraction_ts, forecast_periods)
    availability_forecast = forecast_arima(availability_ts, forecast_periods)

    
    water_depth_forecast = extraction_forecast - availability_forecast

    
    forecast_df = pd.DataFrame({
        "Year": future_years,
        "Extraction Forecast (%)": extraction_forecast,
        "Availability Forecast (ham)": availability_forecast,
        "Forecasted Water Table Depth (m)": water_depth_forecast
    })

    print("\nFuture Forecast Values:")
    print(forecast_df)

    
    sns.set_style("whitegrid")
    plt.figure(figsize=(16, 8))

    sns.lineplot(x=forecast_df['Year'], y=forecast_df['Forecasted Water Table Depth (m)'], marker='o', linestyle='--', color='green', label="Forecasted Water Table Depth")
    sns.lineplot(x=forecast_df['Year'], y=forecast_df['Extraction Forecast (%)'], marker='s', linestyle='-', color='red', label="Extraction Forecast (%)")
    sns.lineplot(x=forecast_df['Year'], y=forecast_df['Availability Forecast (ham)'], marker='^', linestyle='-', color='blue', label="Availability Forecast (ham)")

    plt.xlabel("Year")
    plt.ylabel("Values")
    plt.title("Forecasted Ground Water Trends")
    plt.legend()
    plt.grid(True)
    plt.show()

    
    plt.figure(figsize=(16, 8))
    sns.kdeplot(extraction_forecast, label="Extraction Distribution", fill=True, color='red', alpha=0.4)
    sns.kdeplot(availability_forecast, label="Availability Distribution", fill=True, color='blue', alpha=0.4)
    plt.title("Distribution of Ground Water Extraction vs Availability Forecast")
    plt.xlabel("Forecasted Values")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    
    fig, ax1 = plt.subplots(figsize=(16, 8))
    ax2 = ax1.twinx()
    ax1.plot(forecast_df['Year'], forecast_df['Extraction Forecast (%)'], 'r-', label="Extraction Forecast (%)")
    ax2.plot(forecast_df['Year'], forecast_df['Forecasted Water Table Depth (m)'], 'g-', label="Forecasted Water Table Depth")
    ax1.set_xlabel("Year")
    ax1.set_ylabel("Extraction Forecast (%)", color='red')
    ax2.set_ylabel("Water Table Depth (m)", color='green')
    plt.title("Forecasted Trends with Dual Y-Axis")
    fig.tight_layout()
    plt.show()

    return forecast_df





FIELDS = [
    'Stage of Ground Water Extraction (%)',
    'Net Annual Ground Water Availability for Future Use (ham))',
    'Rainfall(mm)'
]


def evaluate(individual):
    extraction, availability, rainfall = individual  

    
    availability = max(1, availability)

    
    over_extraction = max(0, extraction - availability)  

    
    sustainability_score = max(0, 100 - (over_extraction / availability * 100))  

    return (sustainability_score,)  


def assess_risk(score):
    if score >= 80:
        return "Low", "Sustainable water management achieved. Maintain current policies and focus on conservation."
    elif 50 <= score < 80:
        return "Medium", "Moderate risk. Improve water conservation and regulate groundwater use."
    else:
        return "High", "Urgent action required. Implement strict water regulations, increase recharge, and invest in alternative water sources. Plant trees like Neem, Ashoka, Tulsi, Banana, etc."


creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)

toolbox = base.Toolbox()
toolbox.register("attr_float", random.uniform, 0, 100)
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_float, n=len(FIELDS))
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

toolbox.register("evaluate", evaluate)
toolbox.register("mate", tools.cxBlend, alpha=0.5)
toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=5, indpb=0.2)
toolbox.register("select", tools.selTournament, tournsize=3)


pop = toolbox.population(n=20)
NGEN, CXPB, MUTPB = 50, 0.5, 0.2

for gen in range(NGEN):
    offspring = algorithms.varAnd(pop, toolbox, cxpb=CXPB, mutpb=MUTPB)
    fits = list(map(toolbox.evaluate, offspring))
    for ind, fit in zip(offspring, fits):
        ind.fitness.values = fit
    pop = toolbox.select(offspring, k=len(pop))


best_individual = tools.selBest(pop, k=1)[0]
best_score = evaluate(best_individual)[0]
risk_level, suggestion = assess_risk(best_score)


print("\nFinal Risk Assessment and Suggestions:")
print(f"Overall Risk Level: {risk_level}")
print(f"Final Suggestion: {suggestion}\n")

print("🌿 Best Sustainable Water Policy Found:")
for i, field in enumerate(FIELDS):
    print(f"{field}: {best_individual[i]}")
print(f"🌱 Sustainable Score: {round(best_score, 2)} %")





import requests
import math
from PIL import Image, ImageOps
from io import BytesIO
import matplotlib.pyplot as plt
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
import urllib3

urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

TILE_SIZE = 0.17578125
ASCII_CHARS = "@%#*+=-:. "

STATE_LAYERS = {
    "bihar": "gw:BR_LGEOM",
    "punjab": "gw:PB_LGEOM",
    "uttar pradesh": "gw:UP_LGEOM",
    "up": "gw:UP_LGEOM",
    "odisha": "gw:OD_LGEOM",
    "orissa": "gw:OD_LGEOM",
    "karnataka": "gw:KA_LGEOM",
    "maharashtra": "gw:MH_LGEOM",
    "mh": "gw:MH_LGEOM",
    "gujarat": "gw:GJ_LGEOM",
    "tamil nadu": "gw:TN_LGEOM",
    "tn": "gw:TN_LGEOM",
    "delhi": "gw:DL_LGEOM",
    "nct": "gw:DL_LGEOM"
}

HEADERS = {
    "User-Agent": "GroundwaterViewer/1.0 (Rama Society of Research and Development@sample.com)"  
}

geolocator = Nominatim(user_agent="groundwater_app")


def snap_bbox(lat, lon):
    min_lon = math.floor(lon / TILE_SIZE) * TILE_SIZE
    min_lat = math.floor(lat / TILE_SIZE) * TILE_SIZE
    max_lon = min_lon + TILE_SIZE
    max_lat = min_lat + TILE_SIZE
    return min_lon, min_lat, max_lon, max_lat


def get_state_layer(state_name):
    return STATE_LAYERS.get(state_name.strip().lower())


def fetch_groundwater_tile(lat, lon, layer):
    min_lon, min_lat, max_lon, max_lat = snap_bbox(lat, lon)
    url = "https://bhuvan-vec1.nrsc.gov.in/bhuvan/gwc/service/wms"
    params = {
        "LAYERS": layer,
        "TRANSPARENT": "TRUE",
        "SERVICE": "WMS",
        "VERSION": "1.1.1",
        "REQUEST": "GetMap",
        "STYLES": "",
        "FORMAT": "image/png",
        "SRS": "EPSG:4326",
        "BBOX": f"{min_lon},{min_lat},{max_lon},{max_lat}",
        "WIDTH": "256",
        "HEIGHT": "256"
    }
    print("🛰️ Fetching groundwater map image from Bhuvan...")
    response = requests.get(url, params=params, headers=HEADERS, verify=False)
    response.raise_for_status()
    return Image.open(BytesIO(response.content))


def get_groundwater_value(lat, lon, layer):
    min_lon, min_lat, max_lon, max_lat = snap_bbox(lat, lon)
    x, y = 128, 128
    url = "https://bhuvan-vec1.nrsc.gov.in/bhuvan/gwc/service/wms"
    params = {
        "SERVICE": "WMS",
        "VERSION": "1.1.1",
        "REQUEST": "GetFeatureInfo",
        "LAYERS": layer,
        "QUERY_LAYERS": layer,
        "STYLES": "",
        "SRS": "EPSG:4326",
        "BBOX": f"{min_lon},{min_lat},{max_lon},{max_lat}",
        "WIDTH": "256",
        "HEIGHT": "256",
        "FORMAT": "image/png",
        "INFO_FORMAT": "application/json",
        "X": str(x),
        "Y": str(y)
    }
    print("📊 Fetching groundwater value from Bhuvan...")
    try:
        response = requests.get(url, params=params, headers=HEADERS, verify=False)
        response.raise_for_status()
        data = response.json()

        features = data.get("features")
        if features:
            properties = features[0].get("properties", {})
            print("📈 Groundwater data at location:")
            for k, v in properties.items():
                print(f"  {k}: {v}")
        else:
            print(" No feature data returned.")
    except Exception as e:
        print(" Failed to fetch groundwater value:", e)


def show_image(img):
    plt.imshow(img)
    plt.axis("off")
    plt.title("Groundwater Map Tile")
    plt.show()


def image_to_ascii(img):
    img = img.convert("L")
    img = ImageOps.invert(img)
    img = img.resize((80, 40))
    pixels = img.getdata()
    ascii_image = ''.join(
        ASCII_CHARS[min(pixel * len(ASCII_CHARS) // 256, len(ASCII_CHARS) - 1)]
        for pixel in pixels
    )
    ascii_lines = '\n'.join(ascii_image[i:i + 80] for i in range(0, len(ascii_image), 80))
    print("\n🖼️ ASCII Art Preview:\n")
    print(ascii_lines)



def get_location_from_user():
    lat_input = input("Enter latitude (or press Enter to use your current location): ")
    lon_input = input("Enter longitude (or press Enter to use your current location): ")

    if lat_input and lon_input:
        return float(lat_input), float(lon_input)
    else:
        
        print("📍 Using your current location: Latitude=30.912, Longitude=75.8538")
        return 30.912, 75.8538



def get_state_from_coordinates(lat, lon):
    try:
        location = geolocator.reverse((lat, lon), timeout=10)
        if location and location.raw.get("address", {}).get("state"):
            state = location.raw["address"]["state"]
            print(f"🗺️ Detected state from coordinates: {state}")
            return state
        else:
            print("⚠️ Unable to determine state from coordinates.")
            return None
    except GeocoderTimedOut:
        print("⏳ Geocoding service timed out.")
        return None


def main():
    try:
        lat, lon = get_location_from_user()

        state = get_state_from_coordinates(lat, lon)
        if not state:
            print("Could not determine state. Please enter manually.")
            state = input("Enter your state (e.g., Bihar, UP, Odisha): ").strip()

        layer = get_state_layer(state)
        if not layer:
            print(" Unsupported state. Available options:")
            print(', '.join(sorted(set(key.title() for key in STATE_LAYERS))))
            return

        filename = input("Enter filename to save image (default: groundwater_map.png): ").strip() or "groundwater_map.png"

        img = fetch_groundwater_tile(lat, lon, layer)
        img.save(filename)
        print(f" Image saved as '{filename}'")

        show_image(img)
        image_to_ascii(img)
        get_groundwater_value(lat, lon, layer)

    except requests.exceptions.RequestException as e:
        print(" Network error:", e)
    except Exception as e:
        print(" Unexpected error:", e)


if __name__ == "__main__":
    main()








    def forecast_arima(series, forecast_periods=15, order=(1, 1, 1)):
        if len(series) < 3:
            # Too little data to forecast meaningfully
            return pd.Series([series.iloc[-1]] * forecast_periods)
    
        try:
            model = ARIMA(series, order=order)
            model_fit = model.fit()
            forecast = model_fit.forecast(steps=forecast_periods)
            return forecast
        except Exception as e:
            print(f"ARIMA error: {e}")
            return pd.Series([series.iloc[-1]] * forecast_periods)

  
    def groundwater_forecast(filtered_df, forecast_periods=15):
        
        filtered_df['Year'] = pd.to_numeric(filtered_df['Year'], errors='coerce')
        filtered_df = filtered_df.dropna(subset=['Year'])
    
        years = sorted(filtered_df['Year'].unique())
        future_years = list(range(int(years[-1]) + 1, int(years[-1]) + forecast_periods + 1))
    
        
        extraction_ts = filtered_df.groupby('Year')['Stage of Ground Water Extraction (%)'].mean()
        availability_ts = filtered_df.groupby('Year')['Net Annual Ground Water Availability for Future Use (ham))'].mean()
    
        
        extraction_forecast = forecast_arima(extraction_ts, forecast_periods)
        availability_forecast = forecast_arima(availability_ts, forecast_periods)
    
      
        if 'Depth to Water Level (m)' in filtered_df.columns:
            base_depth = filtered_df['Depth to Water Level (m)'].mean()
        else:
            base_depth = 10  

        # Simple model: 
        stress_factor = (extraction_forecast / 100) * (1 / (availability_forecast.max() + 1e-6))  
        water_depth_forecast = base_depth + stress_factor.cumsum()
    
        # Combine all forecast data
        forecast_df = pd.DataFrame({
            "Year": future_years,
            "Extraction Forecast (%)": extraction_forecast.values,
            "Availability Forecast (ham)": availability_forecast.values,
            "Forecasted Water Table Depth (m)": water_depth_forecast.values
        })
    
        print("\n✅ Future Forecast Values:")
        print(forecast_df)
    
        return forecast_df


    def plot_forecast(forecast_df):
        plt.figure(figsize=(10, 6))
        plt.plot(forecast_df["Year"], forecast_df["Forecasted Water Table Depth (m)"], marker='o', color='teal', label="Water Table Depth (m)")
        plt.xlabel("Year")
        plt.ylabel("Depth (m)")
        plt.title("Forecasted Water Table Depth (2024–2038)")
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()
    
    
    
    
   
    forecast_df = groundwater_forecast(df, forecast_periods=15)
    
    
    plot_forecast(forecast_df)


    
    sns.set_style("whitegrid")
    plt.figure(figsize=(16, 8))

    sns.lineplot(x=forecast_df['Year'], y=forecast_df['Forecasted Water Table Depth (m)'], marker='o', linestyle='--', color='green', label="Forecasted Water Table Depth")
    sns.lineplot(x=forecast_df['Year'], y=forecast_df['Extraction Forecast (%)'], marker='s', linestyle='-', color='red', label="Extraction Forecast (%)")
    sns.lineplot(x=forecast_df['Year'], y=forecast_df['Availability Forecast (ham)'], marker='^', linestyle='-', color='blue', label="Availability Forecast (ham)")

    plt.xlabel("Year")
    plt.ylabel("Values")
    plt.title("Forecasted Ground Water Trends")
    plt.legend()
    plt.grid(True)
    plt.show()

    
    plt.figure(figsize=(16, 8))
    sns.kdeplot(extraction_forecast, label="Extraction Distribution", fill=True, color='red', alpha=0.4)
    sns.kdeplot(availability_forecast, label="Availability Distribution", fill=True, color='blue', alpha=0.4)
    plt.title("Distribution of Ground Water Extraction vs Availability Forecast")
    plt.xlabel("Forecasted Values")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    
    fig, ax1 = plt.subplots(figsize=(16, 8))
    ax2 = ax1.twinx()
    ax1.plot(forecast_df['Year'], forecast_df['Extraction Forecast (%)'], 'r-', label="Extraction Forecast (%)")
    ax2.plot(forecast_df['Year'], forecast_df['Forecasted Water Table Depth (m)'], 'g-', label="Forecasted Water Table Depth")
    ax1.set_xlabel("Year")
    ax1.set_ylabel("Extraction Forecast (%)", color='red')
    ax2.set_ylabel("Water Table Depth (m)", color='green')
    plt.title("Forecasted Trends with Dual Y-Axis")
    fig.tight_layout()
    plt.show()

   


forecast_df = groundwater_forecast(filtered_df, forecast_periods=15)


print(forecast_df)


plt.figure(figsize=(10, 5))
plt.plot(forecast_df['Year'], forecast_df['Stage of Ground Water Extraction (%)'], marker='o', linestyle='-', label='Forecasted Extraction')
plt.plot(forecast_df['Year'], forecast_df['Net Annual Ground Water Availability for Future Use (ham)'], marker='s', linestyle='-', label='Forecasted Availability')

plt.xlabel('Year')
plt.ylabel('Values')
plt.title('Forecasted Groundwater Trends')
plt.legend()
plt.grid()
plt.show()


plt.figure(figsize=(10, 6))
sns.lineplot(x=forecast_df['Year'], y=forecast_df['Extraction Forecast (%)'], 
             marker='o', linestyle='--', color='red', label="Extraction Forecast (%)")


sns.lineplot(x=forecast_df['Year'], y=forecast_df['Availability Forecast (ham)'], 
             marker='o', linestyle='--', color='blue', label="Availability Forecast (ham)")

plt.xlabel("Year")
plt.ylabel("Values")
plt.title("Future Groundwater Extraction, Availability & Depth Forecast")
plt.legend()
plt.xticks(rotation=45)
plt.show()


```

```python
def run_groundwater_ga():
    import random
    import math

   
    POP_SIZE = 6
    BITS_PER_PARAM = 5  
    NUM_PARAMS = 3
    CHROMOSOME_LENGTH = BITS_PER_PARAM * NUM_PARAMS
    GENERATIONS = 10
    MUTATION_RATE = 0.1

    
    def compute_fitness(individual):
        split = BITS_PER_PARAM
        p1 = int(individual[:split], 2)
        p2 = int(individual[split:2*split], 2)
        p3 = int(individual[2*split:], 2)

        
        return (100 - p1) + p2 + p3

    
    def select(population):
        scores = [compute_fitness(ind) for ind in population]
        total = sum(scores)
        probs = [f / total for f in scores]
        return random.choices(population, weights=probs, k=2)

    
    def crossover(p1, p2):
        point = random.randint(1, CHROMOSOME_LENGTH - 1)
        return p1[:point] + p2[point:], p2[:point] + p1[point:]

    
    def mutate(chrom):
        return ''.join(bit if random.random() > MUTATION_RATE else random.choice('01') for bit in chrom)

    
    def create_population():
        return [''.join(random.choice('01') for _ in range(CHROMOSOME_LENGTH)) for _ in range(POP_SIZE)]

    
    def decode(individual):
        split = BITS_PER_PARAM
        return {
            "Stage of Ground Water Extraction (%)": int(individual[:split], 2),
            "Net Annual Ground Water Availability for Future Use (ham)": int(individual[split:2*split], 2),
            "Rainfall (mm)": int(individual[2*split:], 2)
        }

    
    population = create_population()
    print("Initial Population:")
    for ind in population:
        decoded = decode(ind)
        print(f"{ind} -> Stage of Extraction: {decoded['Stage of Ground Water Extraction (%)']}%, "
              f"Availability: {decoded['Net Annual Ground Water Availability for Future Use (ham)']} ham, "
              f"Rainfall: {decoded['Rainfall (mm)']} mm")

    for gen in range(GENERATIONS):
        new_population = []
        for _ in range(POP_SIZE // 2):
            parent1, parent2 = select(population)
            child1, child2 = crossover(parent1, parent2)
            new_population.extend([mutate(child1), mutate(child2)])

        population = new_population
        best = max(population, key=compute_fitness)
        best_values = decode(best)
        print(f"\nGeneration {gen + 1} Best:")
        print(f"  Chromosome: {best}")
        print(f"  Stage of Ground Water Extraction (%): {best_values['Stage of Ground Water Extraction (%)']}")
        print(f"  Net Annual Ground Water Availability for Future Use (ham): {best_values['Net Annual Ground Water Availability for Future Use (ham)']}")
        print(f"  Rainfall (mm): {best_values['Rainfall (mm)']}")
        print(f"  Fitness Score: {compute_fitness(best)}")


run_groundwater_ga()
# Stage of Ground Water Extraction (%) – lower is better

# Net Annual Ground Water Availability for Future Use (ham) – higher is better

# Rainfall (mm) – higher is better
```
